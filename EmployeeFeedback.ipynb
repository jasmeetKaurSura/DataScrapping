### Data Cleaning 

```python
stop_words = stopwords.words('english')
def clean(df):
# removing all punctuations
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    i=0
    for row in df['review']:
        
        for word in row:
            for ele in word:      
                if ele in punc:
                    row = row.replace(ele, " ")

        text_tokens = word_tokenize(row.lower())
        tokens_without_sw = [word for word in text_tokens if word not in stop_words]
        row= (" ").join(tokens_without_sw)
        df['review'][i] = row

        i += 1
    return df
```

```python
df= clean(df)
df
```

### Sentiment Analysis

```python
import json
from nltk.sentiment.vader import  SentimentIntensityAnalyzer
```

```python
SIA= SentimentIntensityAnalyzer()
Polarity_Score=[]
for row in df['review']:
    x = SIA.polarity_scores(row)
    Polarity_Score.append(x)
sentiment_df= pd.DataFrame(Polarity_Score)
print(sentiment_df.head())

```

```python
SentimentAnalysis_df=pd.concat([df.reset_index(drop=True),sentiment_df],axis=1)
print(SentimentAnalysis_df)
SentimentAnalysis_df['sentiment']= SentimentAnalysis_df['compound'].apply(lambda score: 'positive' if score>=0.01 else 'negative' if score<=-0.01 else 'neutral')
SentimentAnalysis_df['keywords']= SentimentAnalysis_df['review'].apply(lambda words: words.split(" "))
print(SentimentAnalysis_df.head())
SentimentAnalysis_df.to_csv('data.csv', index=False)
```
